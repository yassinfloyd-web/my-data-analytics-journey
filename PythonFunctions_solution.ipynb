````python
# Task 6: Function Composition
# 1. 'celsius_to_fahrenheit' converts Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return celsius * 9/5 + 32

# 2. 'fahrenheit_to_celsius' converts Fahrenheit to Celsius
def fahrenheit_to_celsius(fahrenheit):
    # Using parentheses to ensure correct order of operations
    return (fahrenheit - 32) * 5/9

# Test Task 6
print("Task 6 Test:")
# Convert 0°C (Freezing) and 100°C (Boiling) to Fahrenheit
print(f"0°C is {celsius_to_fahrenheit(0)}°F")
print(f"100°C is {celsius_to_fahrenheit(100)}°F")

# Convert 212°F (Boiling) and 32°F (Freezing) back to Celsius
print(f"212°F is {fahrenheit_to_celsius(212)}°C")
print(f"32°F is {fahrenheit_to_celsius(32)}°C")
print()
````

````python
# Task 7: Combining Functions and Conditionals
# Create a function named 'analyze_temperature' that:
# 1. Takes a temperature value and a unit ('C' or 'F')
# 2. Converts the temperature to the other unit
# 3. Returns a string describing if it's "freezing" (below 0°C), 
#    "cold" (0-15°C), "moderate" (15-25°C), or "hot" (above 25°C)
# Your code here


# Test Task 7
print("Task 7 Test:")
# Test with different temperatures in both Celsius and Fahrenheit
# Print the results
print()
````

````python
# Task 7: Combining Functions and Conditionals

# Helper functions from Task 6
def celsius_to_fahrenheit(c):
    return c * 9/5 + 32

def fahrenheit_to_celsius(f):
    return (f - 32) * 5/9

def analyze_temperature(temp, unit):
    # 1. Normalize input to Celsius for consistent analysis
    if unit.upper() == 'F':
        temp_c = fahrenheit_to_celsius(temp)
        other_temp = temp
        other_unit = 'F'
    else:
        temp_c = temp
        other_temp = celsius_to_fahrenheit(temp)
        # The next line is a safety measure. It says: "If the unit was 'C', label the conversion as 'F'. 
        # If the unit was something else entirely (like an error or an empty string), just default the label to 'C'".
        other_unit = 'F' if unit.upper() == 'C' else 'C' 

    # 2. Determine the description based on Celsius thresholds
    if temp_c < 0:
        description = "freezing"
    elif 0 <= temp_c <= 15:
        description = "cold"
    elif 15 < temp_c <= 25:
        description = "moderate"
    else:
        description = "hot"

    return f"{temp}°{unit.upper()} is {description} (approx. {other_temp:.1f}°{other_unit})"

# Test Task 7
print("Task 7 Test:")
print(analyze_temperature(32, 'F'))   # Freezing
print(analyze_temperature(10, 'C'))   # Cold
print(analyze_temperature(72, 'F'))   # Moderate
print(analyze_temperature(30, 'C'))   # Hot
print()
````

### **Normalization Logic**

This section "normalizes" our data. In Data Science, we often convert different formats into a single standard before analyzing them.

* **Normalization:** We ensure `temp_c` always holds the Celsius value, regardless of the input unit.
* **Flexibility:** Using `.upper()` makes the function case-insensitive.
* **Ternary Operator:** We use a one-line `if/else` to determine the label for our converted temperature.

**Instructor Note:** Normalizing your data early in a function prevents you from having to write duplicate logic for different units!

````python
# Task 5: Function with Variable Arguments
# Create a function named 'calculate_average' that accepts any number of numeric arguments
def calculate_average(*numbers):
    # *numbers gathers all arguments into a tuple
    if not numbers:
        return 0
    
    total = sum(numbers)
    count = len(numbers)
    return total / count

# Test Task 8
print("Task 8 Test:")
# Call with different sets of numbers
print(f"Average of (10, 20, 30): {calculate_average(10, 20, 30)}")
print(f"Average of (5, 15): {calculate_average(5, 15)}")
print(f"Average of no numbers: {calculate_average()}")
print()
````

### **Working with Variable Arguments (`*args`)**

Sometimes in Data Science, you don't know how many data points you will receive. Python handles this using the `*` syntax.

* **Flexibility:** Using `*args` (or `*numbers`) allows a function to accept a dynamic number of inputs.
* **Tuple Packing:** Inside the function, the arguments are treated as a **tuple**, which we can iterate over or pass to functions like `sum()`.
* **Error Prevention:** Always check if the input list is empty before performing calculations like division to avoid crashing your script.

**Instructor Note:** This is how many advanced Python libraries allow you to pass an unlimited number of parameters to a single method!

````python
# Task 9: Advanced Challenge - Student Grade Function
# Create a function named 'calculate_final_grade' that:
# 1. Takes a student's name and any number of assignment scores
# 2. Calculates the average score
# 3. Determines the letter grade based on the average (A: 90+, B: 80-89, C: 70-79, D: 60-69, F: <60)
# 4. Returns a formatted string with the student's name, average score, and letter grade
# Your code here


# Test Task 9
print("Task 9 Test:")
# Test with different student names and scores
# Print the results
print()
````

````python
# Task 9: Advanced Challenge - Student Grade Function
def calculate_final_grade(student_name, *scores):
    # 1. Handle cases with no scores to avoid division by zero
    if not scores:
        return f"Student {student_name} has no scores recorded."
    
    # 2. Calculate the average
    average_score = sum(scores) / len(scores)
    
    # 3. Determine the letter grade using conditional logic
    if average_score >= 90:
        letter_grade = 'A'
    elif average_score >= 80:
        letter_grade = 'B'
    elif average_score >= 70:
        letter_grade = 'C'
    elif average_score >= 60:
        letter_grade = 'D'
    else:
        letter_grade = 'F'
    
    # 4. Return formatted string (rounded to 1 decimal place)
    return f"Student: {student_name} | Average: {average_score:.1f} | Grade: {letter_grade}"

# Test Task 9
print("Task 9 Test:")
print(calculate_final_grade("Brittney", 95, 88, 92)) # Test with multiple scores
print(calculate_final_grade("Anura", 70, 85, 78))    # Test with different scores
print(calculate_final_grade("Milton", 100, 98))      # Test with instructor scores
print()
````
