````python
# Solution
# Task 3: if-elif-else Statement
# Write a function that converts a numeric grade to a letter grade
def convert_to_letter_grade(score):
    # We start from the highest score and work our way down
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "f"
    else:
        # If none of the above are True, it must be below 60
        return "F"

# Test Task 3
print("Task 3 Test:")
print(f"85 points is a {convert_to_letter_grade(85)} grade")
print(f"92 points is a {convert_to_letter_grade(92)} grade")
print(f"75 points is a {convert_to_letter_grade(75)} grade")
print(f"65 points is a {convert_to_letter_grade(65)} grade")
print(f"55 points is a {convert_to_letter_grade(55)} grade")
print()
````

### **Key Concepts for Your Students**

*   **Logical Order Matters:** In Python, conditions are evaluated from top to bottom. If we started with `score >= 60`, a student with a 95 would incorrectly receive a "D" because 95 is indeed greater than 60. Starting at 90 ensures the most specific condition is met first.
*   **The `return` Keyword:** Unlike the previous exercises where we used `print()`, this function **returns** a value. This allows the result to be used inside the `print(f"...")` strings in your test code.
*   **Efficiency:** Once a condition is met (e.g., a score of 85 hits the `score >= 80` block), Python ignores all the remaining `elif` and `else` statements for that run.
*   **Readability Counts:** Notice the clean indentation. Keeping the thresholds vertically aligned makes it easy to verify that every grade bracket (90, 80, 70, 60) is covered.

### **The F-String Expression**
The entire string `f"85 points is a {convert_to_letter_grade(85)} grade"` is an expression itself.

*   **The `f` prefix:** This tells Python to look for `{}` placeholders.
*   **Interpolation:** Python takes the string value `"B"` from the function call and "interpolates" it into the text.
*   **The Result:** The result of this expression is the single string: `"85 points is a B grade"`.

````python
# Solution
# Task 4: Nested Conditionals
# Write a function that determines whether a user can access a website
def check_access(is_logged_in, is_admin, is_premium):
    # Outer conditional: Check if the user is even logged in first
    if is_logged_in:
        # Inner conditional: Now that we know they are logged in, check roles
        if is_admin:
            print("Welcome admin! Full access granted")
        elif is_premium:
            print("Welcome premium user! Premium content access granted")
        else:
            print("Welcome! Basic access granted")
    else:
        # This belongs to the outer "if", triggered if is_logged_in is False
        print("Please log in to access the site")

# Test Task 4
print("Task 4 Test:")
check_access(False, False, False)  # Output: Please log in to access the site
check_access(True, True, False)    # Output: Welcome admin! Full access granted
check_access(True, False, True)    # Output: Welcome premium user! Premium content access granted
check_access(True, False, False)   # Output: Welcome! Basic access granted
print()
````

### **Key Concepts for Your Students**

*   **The "Gatekeeper" Pattern:** Using an outer `if` statement to check a primary condition (like `is_logged_in`) is a professional way to organize code. It ensures that Python doesn't waste resources checking for admin or premium status if the user hasn't even signed in yet.
*   **Indentation Levels:** In Python, the "depth" of the indentation tells us which code belongs to which block. The inner `if/elif/else` block is only reached if the outer `is_logged_in` condition is `True`.
*   **Booleans as Conditions:** Since `is_logged_in`, `is_admin`, and `is_premium` are already Booleans (`True` or `False`), we don't need to write `if is_logged_in == True:`. We can simply write `if is_logged_in:`, which is more "Pythonic" and follows the rule that "Simple is better than complex".
*   **Hierarchy of Access:** This structure creates a clear hierarchy. Because the `is_admin` check comes before the `is_premium` check, an admin who also happens to have a premium subscription will still be recognized as an admin first.

### **The "Boolean Check" Pattern**

In Python, writing `if is_logged_in:` is the "Pythonic" way to evaluate a variable that already holds a **Boolean** value (`True` or `False`).

#### **1. How it works**
When Python sees an `if` statement, it evaluates the expression following it to see if it results in `True`.

*   **Satisfied Condition:** If the variable `is_logged_in` is already `True`, the condition is satisfied, and the code block runs.
*   **Skipped Block:** If it is `False`, the code block is skipped.

#### **2. Why not write `if is_logged_in == True:`?**
While `if is_logged_in == True:` is technically correct, it is considered redundant in Python.

*   **Redundancy:** Since `is_logged_in` is already a Boolean, comparing it to `True` is like asking, "Is it true that True is True?".
*   **Readability:** Writing `if is_logged_in:` reads more like a natural English sentence ("If logged in..."), which aligns with the *Zen of Python* principle that "Readability counts."
*   **Simple is better than complex:** This shorter syntax keeps your control flow clean and avoids unnecessary comparison operators.

#### **3. The "Truthiness" Concept**
In Python, the `if` statement doesn't just work for Booleans; it checks for the **"truthiness"** of an object.

*   **Implicit check:** When you write `if variable:`, Python checks if that variable is "truthy."
*   **"Zero to Hero" Example:** For your **#dab-apr-2026** students, you can show them that an empty list `[]` or the number `0` is considered `False` (falsy), while a list with data or a non-zero number is `True` (truthy).

By using `if is_logged_in:`, you are following the "Explicit is better than implicit" rule by relying on the variable's inherent Boolean nature rather than adding extra logic.

````python
# Proving "Falsy" values
print(f"Empty list []: {bool([])}")    # Output: False
print(f"The number 0: {bool(0)}")      # Output: False
print(f"None value: {bool(None)}")     # Output: False

# Proving "Truthy" values
print(f"List with data: {bool([1, 2])}") # Output: True
print(f"The number 10: {bool(10)}")      # Output: True
print(f"A string 'Hi': {bool('Hi')}")   # Output: True
````

### **The Modulo Operator (`%`)**

In Python, the `%` symbol is called the **Modulo Operator**. It returns the **remainder** of a division operation.

**Key Example: Even vs. Odd**
Using the logic from our recent exercise:
* `8 % 2` results in `0` (**Even**, because there is no remainder).
* `5 % 2` results in `1` (**Odd**, because 1 is left over).

**Why it matters:** This is a fundamental tool in programming for controlling logic, such as filtering datasets or creating repeating patterns in your code.

### **Understanding Ternary Operators**

A **ternary operator** (also known as a conditional expression) is a concise way to evaluate a condition and assign a value in a single line. It serves as a shortcut for the standard `if-else` block, supporting the goal of writing "more with less".

#### **1. The Anatomy of a Ternary Statement**
The syntax follows a specific sequence that reads almost like an English sentence:

$$value\_if\_true \text{ if } condition \text{ else } value\_if\_false$$

*   **`value_if_true`**: The result assigned if the condition is **True**.
*   **`if condition`**: The logical test being performed.
*   **`else value_if_false`**: The result assigned if the condition is **False**.

#### **2. Practical Example (from image_4770df.png)**
In the slide example, we determine a person's status based on their age:

```python
age = 20
status = "adult" if age >= 18 else "minor"
```

*   **The Condition:** Python checks if `age >= 18`.
*   **The True Path:** Since 20 is greater than 18, it selects `"adult"`.
*   **The Assignment:** The variable `status` is assigned the value `"adult"`.

### **Usage Guide for Students**

*   **"Readability Counts":** Ternary operators align with the *Zen of Python* by making simple assignments cleaner and more direct.
*   **Avoid "Over-Ternarying":** As noted in the slide, while one ternary statement is efficient, nesting them or using too many can make the code difficult to follow.
*   **When to use:** Use these for simple binary choices (A or B). For complex logic with multiple `elif` conditions, a standard block is usually better for clarity.

````python
# Task 6: Combining Conditions
# Write a function that determines if a year is a leap year
def is_leap_year(year):
    # A leap year is divisible by 4
    # EXCEPT for years divisible by 100, UNLESS also divisible by 400
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

# Test Task 6
print("Task 6 Test:")
print(f"2020 is a leap year: {is_leap_year(2020)}")  # Output: True
print(f"2000 is a leap year: {is_leap_year(2000)}")  # Output: True
print(f"1900 is a leap year: {is_leap_year(1900)}")  # Output: False
print(f"2022 is a leap year: {is_leap_year(2022)}")  # Output: False
print(f"1976 is a leap year: {is_leap_year(1976)}")  # Output: True
print()
````

````python
# Task 7: Best Practices - Refactored
def classify_item(item_type, price, in_stock):
    # Determine the category and tier using ternary operators
    category = "electronics" if item_type == "electronics" else "non-electronics"
    tier = "Affordable" if price < 100 else "Premium"
    
    # Check boolean status directly (Pythonic way)
    stock_status = "in stock" if in_stock else "out of stock"
    
    # Return using an f-string for clarity and performance
    return f"{tier} {category} {stock_status}"

# Test Task 7
print("Task 7 Test (Refactored):")
print(classify_item("electronics", 75, True))   # Output: Affordable electronics in stock
print(classify_item("electronics", 150, False)) # Output: Premium electronics out of stock
print(classify_item("clothing", 50, False))    # Output: Affordable non-electronics out of stock
print()
````
